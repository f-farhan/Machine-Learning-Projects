```python
df= df.drop(['Riot',
       'Explosive', 'Narcotics',
       'Smuggling', 'Total', 'Total Cases'],axis=1)
```

#Value Change

```python
tempdf = df.drop('Dacoity',axis= 1)

labels = df['Dacoity'].iloc[:].values

print(tempdf)

print(labels)
print(labels.shape)
```

```python
feature = tempdf.iloc[:,:].values

print(feature)

print(feature.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature,labels, test_size = 0.1, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import numpy as np
from keras.layers import Dense, Activation
from keras.models import Sequential
```

```python
model = Sequential()

# Adding the input layer and the first hidden layer
model.add(Dense(15, activation = 'relu', input_dim = 10))

# Adding the second hidden layer
model.add(Dense(units = 10, activation = 'relu'))

# Adding the third hidden layer
model.add(Dense(units = 8, activation = 'relu'))

# Adding the fourth hidden layer
model.add(Dense(units = 8, activation = 'relu'))

# Adding the fifth hidden layer
model.add(Dense(units = 4, activation = 'relu'))

# Adding the output layer
model.add(Dense(units = 1))
model.compile(optimizer = 'rmsprop', loss = 'mean_squared_error')
model.summary()
```

```python

# Fitting the ANN to the Training set
history = model.fit(X_train, y_train, batch_size = 162, epochs = 1500)

y_pred = model.predict(X_test)
```

#Value Change

```python
plt.plot(y_test, color = 'red', label = 'Real data')
plt.plot(y_pred, color = 'blue', label = 'Predicted data')
plt.grid()
plt.title('Real Data vs Predicted Data for Dacoity')
plt.legend()
plt.show()
```

```python
print(history.history.keys())

plt.plot(history.history['loss'])
```

#**ANN** Accuracy

```python
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("r2_score",metrics.r2_score(y_test, y_pred))
```

```python


# Model Accuracy: how often is the classifier correct?
print("mean_absolute_error",metrics.mean_absolute_error(y_test, y_pred))
```

```python

# Model Accuracy: how often is the classifier correct?
print("explained_variance_score",metrics.explained_variance_score(y_test, y_pred))
```
